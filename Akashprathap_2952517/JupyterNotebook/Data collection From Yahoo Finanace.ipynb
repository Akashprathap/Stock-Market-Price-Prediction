## Collection of Data from yahoo Finance - 10 YEARS

### 1.CANARA BANK

```python
#importing of libraries
import pandas
from pandas_datareader import data as pdr
import yfinance as yf 
#yfinance aimes to solve this problem by offering a reliable, threaded, and Pythonic way to download historical market data from Yahoo! finance.

import datetime
yf.pdr_override() 

import warnings
warnings.filterwarnings('ignore')

stocks = ["CANBK.BO"]
start = datetime.datetime(2010,1,4)
end = datetime.datetime(2020,1,1)

data = pdr.get_data_yahoo(stocks, start=start, end=end)

#file saved in the directory
data.to_csv('D://Dataset//CANBK.BO.csv') 

data.head(5)
```
